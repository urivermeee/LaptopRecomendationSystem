```python
import re
def process_product_id(product_name, product_brand, product_id):
    # Xóa các từ bắt đầu bằng 'R', 'i', 'Ultra', 'Intel'
    patterns = ['R\d', 'i\d', 'Ultra', 'Intel']
    for pattern in patterns:
        product_name = re.sub(r'\b' + pattern + r'\b.*', '', product_name)
    product_name = re.sub(r'\s+', ' ', product_name).strip()
    # Lấy phần sau dấu "-" 
    parts = product_name.split("-")
    if len(parts) > 1:
        product_name = parts[-1].strip()
    if product_brand == 'Asus':
        return product_name
    elif product_brand == 'MSI':
        return product_name
    elif product_brand == 'Gigabyte':
        return product_name
    else:
        return product_id
    
dt['product_id'] = dt.apply(lambda row: process_product_id(row['product_name'], row['product_brand'], row['product_id']), axis=1)
dt
```

```python
def convert_to_kg(weight):
    # Kiểm tra xem weight có chứa đơn vị g hay không
    weights = weight.replace(' kg', '').replace(' g', '')
    if ' g' in weight:
        # Chuyển đổi từ gram sang kilogram và trả về kết quả
        return float(weights) / 1000
    else:
        # Nếu không chứa đơn vị g, trả về weight ban đầu
        return float(weights)

dt['product_weight'] = dt['product_weight'].apply(convert_to_kg)

bins = [0, 1.5, 2.3, float('inf')]
labels = ['Nhẹ', 'Vừa', 'Nặng']
# Áp dụng bins và labels vào cột 'product_weight_bin'
dt['product_weight_bin'] = pd.cut(dt['product_weight'], bins=bins, labels=labels)
dt['product_weight'] = dt['product_weight'].astype(str) + ' kg'
print(dt[['product_weight', 'product_weight_bin']])
```

```python
def categorize_laptop(cpu, card, ram, scanning, hard_driver, rambus, weight, brand, chip):
    if (
        (cpu in ['Intel Core i5', 'Intel Core i7', 'Intel Core i9', 'Intel Ultra Series', 'AMD Ryzen 5', 'AMD Ryzen 7', 'AMD Ryzen 9']) and
        (('Có Card đồ họa chuyên dụng' in card) or (chip == 'Tích hợp bộ xử lý đồ họa thế hệ mới') ) and
        (ram != '4 GB') and
        (scanning in ['Cao', 'Rất cao']) and
        (hard_driver in ['SSD 512 GB', 'SSD 1 TB', 'SSD 2 TB']) and
        (rambus in ['Cao', 'Trung bình'])
    ):
        return 'Gaming, Đồ họa - Kỹ thuật'
    elif (
        (cpu in ['Intel Core i3', 'Intel Core i5', 'Intel Core i7', 'AMD Ryzen 3', 'AMD Ryzen 5', 'AMD Ryzen 7']) and
        (hard_driver in ['SSD 128 GB', 'SSD 256 GB', 'SSD 512 GB'])
    ):
        return 'Văn phòng, Học tập'
    elif brand == 'Apple':
        return 'MacBook'
    elif weight == 'Nhẹ':
        return 'Mỏng nhẹ'
    else:
        return 'Khác'

# Áp dụng hàm cho mỗi hàng trong DataFrame và gán kết quả vào cột 'categorize_laptop'
dt['categorize_laptop'] = dt.apply(lambda row: categorize_laptop(row['CPU_filter'], row['card_num'], row['RAM_capacity'],
                                                   row['screen_scanning_freq_bin'], row['hard_driver'],
                                                   row['RAM_bus_filtered'], row['product_weight_bin'],row['product_brand'], row['CPU_chip_type']), axis=1)
dt
```

Thêm cột 'discount_percentage' để xác định % giảm giá của sản phẩm

```python
def fpt_calculate_discount_percentage(row):
    discounted_price = float(row['discounted_price'].replace('₫', '').replace('.',''))
    original_price = float(row['original_price'].replace('₫', '').replace('.',''))
    discount_percentage = round(((1 - (discounted_price / original_price)) * 100), 1)
    return discount_percentage

# Áp dụng hàm cho mỗi hàng trong DataFrame và gán kết quả vào cột 'discount_percentage'
fpt['discount_percentage'] = fpt.apply(fpt_calculate_discount_percentage, axis=1)
fpt['discount_percentage'] = fpt['discount_percentage'].astype(str) + '%'

# In ra cột 'discount_percentage' để kiểm tra
print(fpt['discount_percentage'])
```
